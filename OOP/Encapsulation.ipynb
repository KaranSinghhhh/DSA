```python
class BasePlayer:
    def __init__(self, hp, mana):
        self.__hp = hp
        self.__mana = mana

    def get_hp(self):
        return self.__hp
    
    def set_hp(self, hp):
        if hp < 0:
            return "HP must be greater than 0"
        else:
            self.__hp = hp

    def get_mana(self):
        return self.__mana
    
    def set_mana(self, mana):
        if mana < 0:
            return "Mana must be greater than 0"
        else:
            self.__mana = mana

class Wizard(BasePlayer):
    def __init__(self, hp, mana, magic):
        super().__init__(hp, mana)
        self.__magic = magic

    def get_magic(self):
        return self.__magic
    
    def set_magic(self, magic):
        if magic == "Lightning":
            return "You are strong"
        elif magic == "Fire":
            return "You are semi-strong"
        else:
            self.__magic = magic
            return f"Your magic is set to default: {self.__magic}"
        
class Archer(BasePlayer):
    def __init__(self, hp, mana, arrows):
        super().__init__(hp, mana)
        self.__arrows = arrows

    def get_arrows(self):
        return self.__arrows
    
    def set_arrows(self, arrows):
        if arrows < 0:
            return "HP must be greater than 0"
        else:
            self.__arrows = arrows

# Creating instances of Wizard and Archer classes
wizard = Wizard(500, 100, "Fire")
print(wizard.get_hp())
print(wizard.get_mana())
print(wizard.get_magic())

print("--------------------------")

wizard.set_hp(450)
wizard.set_mana(90)
print(wizard.get_hp())
print(wizard.get_mana())
print(wizard.set_magic("Lightning"))
print(wizard.get_magic())


print("--------------------------")

archer = Archer(100, 50, 10)
print(archer.get_hp())
print(archer.get_mana())
print(archer.get_arrows())

print("--------------------------")

archer.set_hp(90)
archer.set_mana(40)
archer.set_arrows(8)
print(archer.get_hp())
print(archer.get_mana())
print(archer.get_arrows())
```

Or we can do this:

```python
class BasePlayer:
    def __init__(self, hp, mana):
        self.__hp = hp
        self.__mana = mana

    @property
    def hp(self):
        return self.__hp
    
    @hp.setter
    def hp(self, value):
        if value < 0:
            print("HP must be greater than 0")
        else:
            self.__hp = value

    @property
    def mana(self):
        return self.__mana
    
    @mana.setter
    def mana(self, value):
        if value < 0:
            print("Mana must be greater than 0")
        else:
            self.__mana = value

class Wizard(BasePlayer):
    def __init__(self, hp, mana, magic):
        super().__init__(hp, mana)
        self.__magic = magic

    @property
    def magic(self):
        return self.__magic

    @magic.setter
    def magic(self, value):
        if value == "Lightning":
            self.__magic = value
            print("You are strong")
        elif value == "Fire":
            self.__magic = value
            print("You are semi-strong")
        else:
            self.__magic = value
            print(f"Your magic is set to default: {self.__magic}")
        
class Archer(BasePlayer):
    def __init__(self, hp, mana, arrows):
        super().__init__(hp, mana)
        self.__arrows = arrows

    @property
    def arrows(self):
        return self.__arrows
    
    @arrows.setter
    def arrows(self, value):
        if value < 0:
            print("Arrows must be greater than 0")
        else:
            self.__arrows = value

# Creating instances of Wizard and Archer classes
wizard = Wizard(500, 100, "Fire")
print(wizard.hp)  # Accessing hp property
print(wizard.mana)  # Accessing mana property
print(wizard.magic)  # Accessing magic property

print("--------------------------")

wizard.hp = -1  # Setting hp using property setter
wizard.mana = -1  # Setting mana using property setter
print(wizard.hp)  # Accessing hp property
print(wizard.mana)  # Accessing mana property
print(wizard.magic)  # Accessing magic property
wizard.magic = "Lightning"  # Setting magic using property setter
print(wizard.magic)  # Accessing magic property

print("--------------------------")

archer = Archer(100, 50, 10)
print(archer.hp)  # Accessing hp property
print(archer.mana)  # Accessing mana property
print(archer.arrows)  # Accessing arrows property

print("--------------------------")

archer.hp = 90  # Setting hp using property setter
archer.mana = 40  # Setting mana using property setter
archer.arrows = 8  # Setting arrows using property setter
print(archer.hp)  # Accessing hp property
print(archer.mana)  # Accessing mana property
print(archer.arrows)  # Accessing arrows property
```
