```python
class Archer:
    def __init__(self, hp, mana, arrows): 
        self.hp = hp 
        self.mana = mana  
        self.arrows = arrows 

    def __str__(self):
        return f"Archer with {self.hp} HP and {self.mana} mana and {self.arrows} arrows"
    
    #def __repr__(self):
    #   return f"Archer({self.hp}, {self.mana}, {self.arrows})"

    def __add__(self, other): #other is another instance of the Archer class
        new_hp = self.hp + other.hp
        new_mana = self.mana + other.mana
        new_arrows = self.arrows + other.arrows
        return Archer(new_hp, new_mana, new_arrows)
    
    def __eq__(self, other):
        if self.hp == other.hp and self.mana == other.mana and self.arrows == other.arrows:
            return True
        else:
            return False
            
    
    def __gt__(self, other):
        if self.hp > other.hp and self.mana > other.mana and self.arrows > other.arrows:
            return True
        else:
            return False
        
    def __lt__(self, other):
        if self.hp < other.hp and self.mana < other.mana and self.arrows < other.arrows:
            return True
        else:
            return False
        
    def __hash__(self):
        return hash((self.hp, self.arrows, self.mana))

archer1 = Archer(100, 100, 5)
archer2 = Archer(100, 100, 5) #INSTANCE of Archer class

#When you use the "+"" operator with instances of the Archer class, Python internally calls the __add__ method.
archer3 = archer1 + archer2 #Here, archer1 is self, and archer2 is other within the "__add__" magic method
print(archer3)

#when we use the "=" operator with instances of the Archer class, python internally calls the __eq__ method
print(archer1 == archer2) #Here, archer1 is self, and archer2 is other within the "__eq__" magic method

#">" and "<" operators:
print(archer1 > archer2)
print(archer2 > archer1)
print(archer3 > archer2)
print(archer3 < archer2)

# Adding instances to a set (which uses hashes)
archers_set = {archer1, archer2, archer3}
print(len(archers_set))  # Output: 2, because archer1 and archer2 are considered equal

# Using instances as dictionary keys
archers_dict = {archer1: "First archer", archer3: "Second archer"}
print(archers_dict[archer2])  # Output: "First archer", because archer1 and archer2 are considered equal
```
